```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
X = 2*np.random.rand(100,1)
y = 6 + 4*X + np.random.randn(100,1)
plt.scatter(X,y)
```

```python
def gradient_descent(X,y,iters=10000):
    w0 = np.zeros((1,1))
    w1 = np.zeros((1,1))
    
    for _ in range(iters):
        N = len(y)
        w0_update = np.zeros_like(w0)
        w1_update = np.zeros_like(w1)
        y_pred = np.dot(X,w1.T) + w0
        diff = y - y_pred
        w0_factors = np.ones((N,1))
        w0_update = -(2/N)*0.01*(np.dot(w0_factors.T,diff))
        w1_update = -(2/N)*0.01*(np.dot(X.T,diff))
        w0 = w0 - w0_update
        w1 = w1 - w1_update
    return w1,w0,y_pred
```

```python
w1,w0,y_pred = gradient_descent(X,y,iters=1000)
print(w1,w0)
print(np.sum(np.square(y-y_pred))/len(y))
```
